```python
# Drop the currency column, as all of them contain the same value
df = df.drop('currency', axis='columns')
```

```python
# Create helper function to represent the total amount of bytes in Mb.
def _byte_resizer(data):
    return np.around(data / 1000000, decimals=2)
```

```python
# Create a new series representing the bytes in Mb using our custom function
df['size_Mb'] = df['size_bytes'].apply(_byte_resizer)
```

```python
# Drop the size_bytes column from our current dataframe
df.drop('size_bytes', axis='columns', inplace=True)
```

```python
df['size_Mb']
```

```python
# Generates a subplot with our x and y axis defined
plt.subplots(figsize=(10, 8))
```

```python
# Create the bins and labels for the data
BINS = [0.00, 10.00, 20.00, 50.00, 100.00, 200.00, 500.00, 1000.00, 2000.00, np.inf]
LABELS = ["<10m", "10-20m", "20-50m", "50-100m", "100-200m", "200-500m", "500-1000m", "1-2G", ">2G"]
```

```python
# Segment our data into the bins we have created
frequencies = pd.cut(df['size_Mb'], BINS, include_lowest=True, labels=LABELS)
```

```python
# Plot the data using the segmented data and labels we specified
# Need to remove null data values, as they are currently affecting results
sb.barplot(y=frequencies.value_counts(), x=frequencies.value_counts().index)
```

```python
BINS = [-np.inf, 0, np.inf]
LABELS = ['FREE', 'PAID']
colors = ['lightcoral', 'yellowgreen']
```

```python
# Create a new column that contains free vs. paid attributes
df['price_categories'] = pd.cut(df['price'], BINS, include_lowest=True, labels=LABELS)
```

```python
# Get the
price_df = df['price_categories'].value_counts()
```

```python
# Create a pie chart and a circle for 
plt.pie(price_df.values, labels=LABELS, colors=colors, autopct='%1.1f%%', shadow=True)
center_circle = plt.Circle((0,0), 0.75, color='black', fc='white', linewidth=1.25)

# Get the current figure and then add our circle to it's axes
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Set the axis properties to be equal, which turns of axis lines and labels
plt.axis('equal')
```

```python
# Slice our price categories into two separate dataframes (free and paid for)
free_apps = df.loc[df['price_categories'] == 'FREE']
paid_apps = df.loc[df['price_categories'] == 'PAID']
```

```python
# From our free and paid for apps, create a new sorted dataframe where the apps are
# sorted descendingly by their ratings
free_rated_apps = free_apps.sort_values(by=['rating_count_tot'], ascending=False)
paid_rated_apps = paid_apps.sort_values(by=['rating_count_tot'], ascending=False)
```

```python
# Plot the top rated free apps
sb.barplot(x=free_rated_apps['rating_count_tot'][:10], y=free_rated_apps['track_name'][:10])
```

```python
sb.barplot(x=paid_rated_apps["rating_count_tot"][:10], y=paid_rated_apps["track_name"][:10])
```

```python
# Get all the ratings in a new df sorted by the total ratings
ratings = df.sort_values(by=['rating_count_tot'], ascending=False)
```

```python
# Plot the top rated apps
plt.subplots(figsize=(20, 20))
sb.barplot(x=ratings['rating_count_tot'][:30], y=ratings['track_name'][:30])
```

```python
# Get all the genres and then sort them in descending order
genres = df['prime_genre'].value_counts()
genres.sort_values(ascending=False, inplace=True)

# Initialize plot and then draw our genre values on it
plt.subplots(figsize=(20, 10))
sb.barplot(x=genres.values, y=genres.index, order=genres.index, orient='h')
```

```python
# Create a dataframe giving us all apps that are within the games genre
games = df.loc[df['prime_genre'] == 'Games']
games.head()
```

```python
prices = (games['price'].value_counts()) / (games['price'].shape[0]) * 100
prices.sort_values(ascending=False, inplace=True)
```

```python
# plot the average prices of 
plt.subplots(figsize=(20, 10))
ax = sb.barplot(y=prices.values, x=prices.index, order=prices.index)
ax.set(xlabel='USD', ylabel='percent (%)')
```

```python
# Get all games that are free & paid for sliced into their own data frames
free_games = games.loc[games['price_categories'] == 'FREE']
paid_games = games.loc[games['price_categories'] == 'PAID']

# Get all top free and paid for games sorted by total rating!
free_rated_games = free_games.sort_values(by=['rating_count_tot'], ascending=False)
paid_rated_games = paid_games.sort_values(by=['rating_count_tot'], ascending=False)
```

```python
# Create our figure and two subplots to render both our top free and paid for games
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# Plot both categories of games!
sb.barplot(x=free_rated_games['rating_count_tot'][:10], y=free_rated_games['track_name'][:10], ax=ax1)
sb.barplot(x=paid_rated_games['rating_count_tot'][:10], y=paid_rated_games['track_name'][:10], ax=ax2)
```

```python
# Get the currently top rated free and paid for games
current_free_rated_games = free_games.sort_values(by=['rating_count_ver'], ascending=False)
current_paid_rated_games = paid_games.sort_values(by=['rating_count_ver'], ascending=False)
```

```python
# Initialize and configure plot
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# Plot the currently highest rated free and paid for games
sb.barplot(x=current_free_rated_games['rating_count_ver'][:10], y=current_free_rated_games['track_name'][:10], ax=ax1)
sb.barplot(x=current_paid_rated_games['rating_count_ver'][:10], y=current_paid_rated_games['track_name'][:10], ax=ax2)
```
